# **Pedra Papel Tesoura!**

The project is about a jokenpô, the famous game of Rock, Paper, Scissors in the Python language. This is a very simple code, used as a basis for the Github features and the student pack, for the code innovation series Brasil




```python
from random import choice
print('Bem-vindo ao jogo de Jokenpô, selecione Pedra, Papel ou Tesoura! \nFaça sua jogada:' )
jogador = str(input('Selecione o numero para: \n1 - Pedra \n2 - Papel \n3 - Tesoura'))
print('JO...KEN...PÔÔÔÔÔÔÔÔÔ!')
PEDRA = '1'
PAPEL = '2'
TESOURA = '3'
opcao = [PEDRA, PAPEL, TESOURA]
computador = choice(opcao)
if computador == PEDRA and jogador == '1':
    print('O jogador escolheu Pedra! \n...\nComputador escolheu Pedra... \n...\nEmpatou!')
    
if computador == PEDRA and jogador == '2':
    print('O jogador escolheu Papel! \n...\nComputador escolheu Pedra... \n...\nVenceu! Papel embrulha a pedra!!!')
    
if computador == PEDRA and jogador == '3':
    print('- O jogador escolheu Tesoura! \n...\n- Computador escolheu Pedra... \n...\nPerdeu! A pedra quebra a tesoura!!!')
   
if computador == PAPEL and jogador == '1':
    print('- O jogador escolheu Pedra! \n...\n- Computador escolheu Papel... \n...\nPerdeu! O Papel embrulha a Pedra!!!')
   
if computador == PAPEL and jogador == '2':
     print('- O jogador escolheu Papel! \n...\n- Computador escolheu Papel... \n...\nEmpatou!')

if computador == PAPEL and jogador == '3':
    print('- O jogador escolheu Tesoura! \n...\n- Computador escolheu Papel... \n...\nVenceu! A Tesoura corta o Papel!!!')
  
if computador == TESOURA and jogador == '1':
    print('O jogador escolheu Pedra! \n...\nComputador escolheu Tesoura... \n...\nVenceu! A Pedra quebra a Tesoura!!!')

if computador == TESOURA and jogador == '2':
    print('O jogador escolheu Papel! \n...\nComputador escolheu Tesoura... \n...\nPerdeu! Tesoura corta o Papel!!!')

if computador == TESOURA and jogador == '3':
     print('O jogador escolheu Tesoura! \n...\nComputador escolheu Tesoura... \n...\nEmpatou!')
```
